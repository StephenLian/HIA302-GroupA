Importing of team common dataset to ensure no loss of data during initial merging of files. File used collected_covid_daily_cases.csv

```python
url1 = "https://raw.githubusercontent.com/StephenLian/HIA302-GroupA/main/collected_covid_daily_cases.csv"
dfcombined_download = requests.get(url1).content
dfcombined = pd.read_csv(io.StringIO(dfcombined_download.decode('utf-8')))
```

```python
dfcombined
```

```python
dfcombined.info()
```

```python
#calculating total of malaysia covid admissions daily
dfcombinedsum = dfcombined.groupby('date')['admitted_covid'].sum().reset_index()
dfcombinedsum
```

```python
#Getting Malaysia total covid cases daily
dfmalaysia = dfcombined.drop(['state', 'state_cases_new','admitted_covid','icu_covid'], axis=1)
dfdaily = dfmalaysia.drop_duplicates()
```

```python
dfdaily
```

```python
mergedhosp_malaysia = pd.merge (dfdaily, dfcombinedsum, how='outer', on = 'date')
mergedhosp_malaysia
```

```python
mergedhosp_malaysia.info()
```

```python
mergedhosp_malaysia['percentage_admitted'] = (mergedhosp_malaysia['admitted_covid'] / mergedhosp_malaysia['malaysia_cases_new']) *100
mergedhosp_malaysia
```

```python

# making the scatter plot on latitude and longitude
# setting color on the basis of country
fig = alt.Chart(mergedhosp_malaysia).mark_point().encode(
  x='malaysia_cases_new',y='admitted_covid')
  
# making the regression line using transform_regression
# function and add with the scatter plot
final_plot = fig + fig.transform_regression('malaysia_cases_new','admitted_covid').mark_line()
final_plot
```

```python
alt.Chart(mergedhosp_malaysia).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
).properties(
    width=150,
    height=150
).repeat(
    row=['malaysia_cases_new', 'admitted_covid', 'date'],
    column=['date', 'admitted_covid', 'malaysia_cases_new']
).interactive()
```

```python
zscore = mergedhosp_malaysia.drop(['date','holiday'], axis=1)

df_zscore = (zscore - zscore.mean())/zscore.std()

# setting color on the basis of country
zsc = alt.Chart(df_zscore).mark_point().encode(
  x='malaysia_cases_new',y='admitted_covid')
# making the regression line using transform_regression
# function and add with the scatter plot
zscoreplot = zsc + zsc.transform_regression('malaysia_cases_new','admitted_covid').mark_line()

zscoreplot
```

```python
base = alt.Chart(dfcombined).mark_point().encode(
    x='state_cases_new:Q',
    y='admitted_covid:Q',
    tooltip=('date','holiday','malaysia_cases_new','state_cases_new','admitted_covid'),
    color='state:N'
).properties(
    width=300,
    height=300
)

chart = alt.hconcat()
for state in ['Selangor', 'Perak', 'Johor']:
    chart |= base.transform_filter(datum.state == state)
chart
```

```python


# Calculate the 7-day moving average of the admissions data
mergedhosp_malaysia["7-day moving average"] = mergedhosp_malaysia["percentage_admitted"].rolling(window=7).mean()

# Create a line chart using Altair
rollingmean = alt.Chart(mergedhosp_malaysia).mark_point().encode(
    x="date",
    y="7-day moving average",
    tooltip=('date','admitted_covid','percentage_admitted','holiday'),
).properties(width=1500)
rollingmean
```
